```python
def scatter1(x, y, s='', c='', avg=None, years=[2011, 2015]):
    '''
    This function will create a scatterplot of two variables, x and y. 
    To make a colormap of another variable, set c = columnname.
    To make point size correspond to another variable, set s = columnname. 
    To add a line of best fit for the xaxis and yaxis variables, set bestfit = "yes".
    '''
    
    #2011 data
    if 2011 in years:
        xaxis2011 = plume2011[x]
        yaxis2011 = plume2011[y]
        if c != '' and s != '':
            coloraxis2011 = plume2011[c]
            sizeaxis2011 = (plume2011[s]/(plume2011[s].max()))*400
        elif c != '':
            sizeaxis2011 = None
            coloraxis2011 = plume2011[c]
            plt.title(str(y)+" vs "+str(x)+" with color representing "+str(c))
        elif s != '':
            coloraxis2011 = None
            sizeaxis2011 = (plume2011[s]/(plume2011[s].max()))*400
            plt.title(str(y)+" vs "+str(x)+" with size representing "+str(s))
        else:
            coloraxis2011 = None
            sizeaxis2011 = None
            plt.title(str(y)+" vs "+str(x))

        plt.scatter(xaxis2011, yaxis2011, s = sizeaxis2011, c = coloraxis2011, marker = 's', cmap = 'plasma', label = '2011 Plume')

    #2015 data
    if 2015 in years:
        xaxis2015 = plume2015[x]
        yaxis2015 = plume2015[y]
        if c != '':
            coloraxis2015 = plume2015[c]
        else:
            coloraxis2015 = None
        if s != '':
            sizeaxis2015 = (plume2015[s]/(plume2015[s].max()))*400
        else:
            sizeaxis2015 = None

        plt.scatter(xaxis2015, yaxis2015, s = sizeaxis2015, c = coloraxis2015, marker = 'o', cmap = 'plasma', label = '2015 Plume')

    if c != '':
        plt.colorbar(label = str(c))
    #plt.title(str(y)+" vs "+str(x)+" with color representing "+str(c)+ " and size representing "+str(s))
    plt.xlabel(str(x))
    plt.ylabel(str(y))
    
    if avg != None:
        plt.hlines(avg, 11750, 13785, colors=['skyblue'], linestyles='dashed', label = 'Median ' + str(y) + ' Value = '+str(avg))
    plt.legend()
```

```python
def median_getter(df, var, min_alt, max_alt):
    '''
    Using all of the WAS data throughout the years, this method calculates the median value of var,
    within the altitude range (min_alt, max_alt)
    '''
    altRange = df[(df['AltP_feet'] > min_alt) & (df['AltP_feet'] < max_alt)]
    med = np.nanmedian(altRange[var])
    return round(med, 1)
```

```python
#plume1 in plume2015, plume2 is plume2011
def plumes_ks(plume1, plume2, keys):
    red = []
    blue = []
    green = []
    for k in keys:
        results = (st.kstest(plume1[k], plume2[k]))
        if results[0]>0.7:
            color = Fore.RED
            red.append(k)
        elif results[0]>0.3:
            color = Fore.BLUE
            blue.append(k)
        else:
            color = Fore.GREEN
            green.append(k)
        print(color+ 'The KS Test results for '+ k + ' is:')
        print(color + str(results))
        print(color + '2015 median, mean: '+str(np.nanmedian(plume1[k])) +', '+ str(np.nanmean(plume1[k])))
        print(color + '2011 median, mean: '+str(np.nanmedian(plume2[k])) +', '+ str(np.nanmean(plume2[k])))
        print()
    return red, blue, green
```

```python
#define function used to fit line to data
def slopeintfunc(x,s1,incpt):
    return s1*np.array(x)+incpt
```

```python
def by_year(dataframe, year):
    df = dataframe['SARP Year'].astype(int) == year
    data = dataframe[df]
    return data
```

```python
def by_alt(dataframe, min_alt, max_alt = 50000):
    altRange = dataframe[(dataframe['AltP_feet'] > min_alt) & (dataframe['AltP_feet'] < max_alt)]
    return altRange
```

```python
def by_latlong(dataframe, minlat, maxlat, minlong, maxlong):
    data = dataframe[(dataframe['Latitude'] > minlat) & (dataframe['Latitude'] < maxlat) & (dataframe['Longitude'] > minlong) & (dataframe['Longitude'] < maxlong)]
    return data
```

```python
def by_flight(dataframe, number):
    df = dataframe['Flight'].astype(str) == number
    data = dataframe[df]
    return data
```

```python
def yearly_trends(compound, min_alt, max_alt = 50000, minlat=30, maxlat=40, minlong=-125, maxlong=-110, min_year = 2009, max_year = 2019, omissions=[], bestfit = 'no'):
    '''
    
    '''
    years = []
    
    backgrounds = []
    
    altRange = full[(full['AltP_feet'] > min_alt) & (full['AltP_feet'] < max_alt)]
    
    for year in range(min_year, max_year+1):
        if year in omissions:
            pass
        else:
            years.append(year)
            data = by_latlong((by_alt(by_year(full, year), min_alt, max_alt)),minlat, maxlat, minlong, maxlong)
            med = np.nanmedian(data[compound])
            backgrounds.append(med)
    plt.scatter(years, backgrounds)
    plt.title('Median Concentrations of '+compound+' in the Altitude Range ('+str(min_alt)+'ft,'+str(max_alt)+'ft) during the years '+str(min_year)+'-'+str(max_year))
    plt.xlabel('Years')
    plt.ylabel(compound)
    plt.xlim(min_year-1, max_year+1) 
    if bestfit == "yes":
        backgrounds = np.array(backgrounds)
        years = np.array(years)
        cleany = backgrounds[(np.isnan(backgrounds)==False) & (np.isnan(years)==False)]
        cleanx = years[(np.isnan(backgrounds)==False) & (np.isnan(years)==False)]
        
        best = optimization.curve_fit(slopeintfunc, cleanx, cleany)
        print("slope, int =",best[0])
        print("(R (correleation coefficient), p-value) =",pearsonr(cleanx, cleany))
        plt.plot(cleanx, slopeintfunc(cleanx, best[0][0], best[0][1]), color = "salmon")
    return (backgrounds, years)
        #best = optimization.curve_fit(slopeintfunc, years, backgrounds)
        #print("slope, int =",best[0])
        #print("(R (correleation coefficient), p-value) =",pearsonr(years, backgrounds))
        #print(best)
        #plt.plot(years, slopeintfunc(years, best[0][0], best[0][1]), color = "salmon")
```
